```python
import os

def rename_md_files(directory):
    try:
        # Recorrer el directorio y subdirectorios
        for root, _, files in os.walk(directory):
            for filename in files:
                # Verificar si el archivo tiene extensiÃ³n .md y contiene un guion bajo
                if filename.endswith(".md") and "_" in filename:
                    # Reemplazar guion bajo por guion
                    new_filename = filename.replace("_", "-")
                    # Obtener las rutas completas
                    old_path = os.path.join(root, filename)
                    new_path = os.path.join(root, new_filename)
                    # Renombrar el archivo
                    os.rename(old_path, new_path)
                    print(f'Renamed: {old_path} -> {new_path}')
        print("Renaming completed.")
    except Exception as e:
        print(f"Error: {e}")

# Ruta de la carpeta que deseas procesar
directory = "./documentation"
rename_md_files(directory)
```
